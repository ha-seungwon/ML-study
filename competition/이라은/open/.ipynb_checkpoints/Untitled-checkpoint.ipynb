```python
def make_timeseries(data, what):
    data_list = []
    
    if what == 'train':
        col1 = data.iloc[:,6:95]
        col2 = data.iloc[:,95:]
        
        tcol1 = col1.T
        tcol2 = col2.T
        
        tcol1.columns = data['ID']
        tcol2.columns = data['ID']+str('_1')
        
        tcol1= tcol1.reset_index().drop('index', axis=1)
        tcol2= tcol2.reset_index().drop('index', axis=1)
        
        for id in data['ID']:
            md = pd.concat([tcol1[f'{id}'], tcol2[f'{id}_1']], axis=1)
            md = md.rename(columns = {f'{id}' : 'ìƒˆìˆœ', f'{id}_1' : 'ì—½ë¡ìˆ˜'})
            md['ID'] = id
            md['ì°©ê³¼ëŸ‰(int)'] = int(data.loc[data['ID'] == id, 'ì°©ê³¼ëŸ‰(int)'])
            md['ìˆ˜ê³ (m)'] = float(data.loc[data['ID'] == id, 'ìˆ˜ê³ (m)'])
            md['ìˆ˜ê´€í­(min)'] = float(data.loc[data['ID'] == id, 'ìˆ˜ê´€í­1(min)'])
            md['ìˆ˜ê´€í­(max)'] = float(data.loc[data['ID'] == id, 'ìˆ˜ê´€í­2(max)'])
            md['ìˆ˜ê´€í­í‰ê· '] = float(data.loc[data['ID'] == id, 'ìˆ˜ê´€í­í‰ê· '])
              
            data_list.append(md)
            
        td = pd.concat(data_list)
        
        return td
        
    if what == 'test':
        col1 = data.iloc[:,5:94]
        col2 = data.iloc[:,94:]
        
        tcol1 = col1.T
        tcol2 = col2.T
        
        tcol1.columns = data['ID']
        tcol2.columns = data['ID']+str('_1')
        
        tcol1= tcol1.reset_index().drop('index', axis=1)
        tcol2= tcol2.reset_index().drop('index', axis=1)
        
        for id in data['ID']:
            md = pd.concat([tcol1[f'{id}'], tcol2[f'{id}_1']], axis=1)
            md = md.rename(columns = {f'{id}' : 'ìƒˆìˆœ', f'{id}_1' : 'ì—½ë¡ìˆ˜'})
            md['ID'] = id
            md['ìˆ˜ê³ (m)'] = float(data.loc[data['ID'] == id, 'ìˆ˜ê³ (m)'])
            md['ìˆ˜ê´€í­(min)'] = float(data.loc[data['ID'] == id, 'ìˆ˜ê´€í­1(min)'])
            md['ìˆ˜ê´€í­(max)'] = float(data.loc[data['ID'] == id, 'ìˆ˜ê´€í­2(max)'])
            md['ìˆ˜ê´€í­í‰ê· '] = float(data.loc[data['ID'] == id, 'ìˆ˜ê´€í­í‰ê· '])
              
            data_list.append(md)
            
        td = pd.concat(data_list)
        
        return td
```

```python
class Predictor(nn.Module):
    def __init__(self, n_features, n_hidden, seq_len, n_layers):
        super(Predictor, self).__init__()
        self.n_hidden = n_hidden
        self.seq_len = seq_len
        self.n_layers = n_layers
        self.lstm = nn.LSTM(
            input_size=n_features,
            hidden_size=n_hidden,
            num_layers=n_layers
        )
        self.linear = nn.Linear(in_features=n_hidden, out_features=1)
    def reset_hidden_state(self):
        self.hidden = (
            torch.zeros(self.n_layers, self.seq_len, self.n_hidden),
            torch.zeros(self.n_layers, self.seq_len, self.n_hidden)
        )
    def forward(self, sequences):
        lstm_out, self.hidden = self.lstm(
            sequences.view(len(sequences), self.seq_len, -1),
            self.hidden
        )
        last_time_step = lstm_out.view(self.seq_len, len(sequences), self.n_hidden)[-1]
        y_pred = self.linear(last_time_step)
        return y_pred
```

```python
def train_model(model, train_data, train_labels, val_data=None, val_labels=None, num_epochs=100, verbose = 10, patience = 10):
    loss_fn = torch.nn.L1Loss() #
    optimiser = torch.optim.Adam(model.parameters(), lr=0.001)
    train_hist = []
    val_hist = []
    for t in range(num_epochs):

        epoch_loss = 0

        for idx, seq in enumerate(train_data): 

            model.reset_hidden_state() # seq ë³„ hidden state reset

            # train loss
            seq = torch.unsqueeze(seq, 0)
            y_pred = model(seq)
            loss = loss_fn(y_pred[0].float(), train_labels[idx]) # 1ê°œì˜ stepì— ëŒ€í•œ loss

            # update weights
            optimiser.zero_grad()
            loss.backward()
            optimiser.step()

            epoch_loss += loss.item()

        train_hist.append(epoch_loss / len(train_data))

        if val_data is not None:

            with torch.no_grad():

                val_loss = 0

                for val_idx, val_seq in enumerate(val_data):

                    model.reset_hidden_state() # seq ë³„ë¡œ hidden state ì´ˆê¸°í™” 

                    val_seq = torch.unsqueeze(val_seq, 0)
                    y_val_pred = model(val_seq)
                    val_step_loss = loss_fn(y_val_pred[0].float(), val_labels[val_idx])

                    val_loss += val_step_loss
                
            val_hist.append(val_loss / len(val_data)) # val histì— ì¶”ê°€

            ## verbose ë²ˆì§¸ ë§ˆë‹¤ loss ì¶œë ¥ 
            if t % verbose == 0:
                print(f'Epoch {t} train loss: {epoch_loss / len(train_data)} val loss: {val_loss / len(val_data)}')

            ## patience ë²ˆì§¸ ë§ˆë‹¤ early stopping ì—¬ë¶€ í™•ì¸
            if (t % patience == 0) & (t != 0):
                
                ## lossê°€ ì»¤ì¡Œë‹¤ë©´ early stop
                if val_hist[t - patience] < val_hist[t] :

                    print('\n Early Stopping')

                    break

        elif t % verbose == 0:
            print(f'Epoch {t} train loss: {epoch_loss / len(train_data)}')

            
    return model, train_hist, val_hist
```
