```python
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
sample=pd.read_csv('sample_submission.csv')
```

```python
train=train.drop('ID',axis=1)
test=test.drop('ID',axis=1)
```

```python
def COMPONENT_ARBITRARY(df):
    temp=df['COMPONENT_ARBITRARY'].tolist()
    input_temp=[]
    for i in temp:
        input_temp.append(int(i[-1]))
    df['COMPONENT_ARBITRARY']=input_temp
    return df

def label_code(dataframe,column):
    
    items=dataframe[column].tolist()
    encoder = LabelEncoder()
    encoder.fit(items)
    dataframe[column]=encoder.transform(items)
    return dataframe
```

```python
train=COMPONENT_ARBITRARY(train)
test=COMPONENT_ARBITRARY(test)
```
